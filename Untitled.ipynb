```python

from pathlib import Path
from typing import Dict
from glob import glob
import os
from itertools import islice
import matplotlib.pyplot as plt
import cv2
```

```python
from Expresssion2RGB import Expression2Img
```

```python
def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))


def validatePath(path: Path)->Path:
    """Validate path"""
    if not path.endswith("/"):
        path = path + "/"
    return path

def getSamplePaths(data_path : Path)->Dict[str, Dict[str, Path]]:
    """Find all data paths for every sample"""
    samplePaths = dict()
    expressionFile = "/filtered_feature_bc_matrix.h5"
    metaFile = "/spatial/tissue_positions_list.csv"
    scalerFile = "/spatial/scalefactors_json.json"
    
    data_path = validatePath(data_path)
    for f in glob(f"{data_path}*"):
        sample_name = os.path.basename(f)
        samplePaths[sample_name] = {"h5_path" : f + expressionFile, 
                                    "spatial_path" : f + metaFile, 
                                    "scale_factor_path" : f + scalerFile}
    return samplePaths
```

```python
samplePaths = getSamplePaths("/sdd/MOB/new_data/")
```

```python
random_sample = take(1, samplePaths.values())[0]
sample = take(1, samplePaths.keys())[0]
```
